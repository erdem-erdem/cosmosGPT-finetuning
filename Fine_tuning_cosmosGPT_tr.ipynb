```python
from transformers import AutoTokenizer, GPT2LMHeadModel
from transformers import pipeline
```

```python
model = GPT2LMHeadModel.from_pretrained("ytu-ce-cosmos/turkish-gpt2")
tokenizer = AutoTokenizer.from_pretrained("ytu-ce-cosmos/turkish-gpt2")

# Huggingface'ten bir model ve tokenizer kullanılır.
```

```python
if tokenizer.pad_token is None:
  tokenizer.pad_token = tokenizer.eos_token
  tokenizer.pad_token_id = tokenizer.eos_token_id
```

Aşağıda çıktı üretimi için gerekl aşamalar gösterilmiştir.

*   Girdi metni, yukarıda yüklenen tokenizer ile tokenlara ayrılır.
*   Bu tokenlar modele verilerek çıktı elde edilir.
*   Üretilen metin tokenlar halindedir. Çıktının metin olarak görülebilmesi için bu tokenların metin karşılıkları elde edilmelidir.

```python
prompt = "Zor günlere rağmen hayatı sevmek için 5 neden sırala."

input_ids = tokenizer(prompt, return_tensors="pt").input_ids

output = model.generate(
    input_ids=input_ids,
    pad_token_id=tokenizer.pad_token_id,
    max_length=100
)

output
```

```python
import torch

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")
```

```python
#Yukarıda gösterilen token dizisi, "decode" fonksiyonu ile metin haline getiliir.

text = tokenizer.decode(output[0], skip_special_tokens=False)
print(text)
```

```python
import datasets
from datasets import load_dataset
from datasets import Dataset as Ds
from torch.utils.data import Dataset
from transformers import DataCollatorForLanguageModeling, Trainer, TrainingArguments
```

```python
ds_len = 2000       #You can experiment with less training to see the result.
                    #We'll just use a tiny bit of the dataset.

                    #Larger numbers give better results but last longer.

dataset = load_dataset("merve/turkish_instructions")["train"].select(range(ds_len))
```

Yukarıda "merve/turkish_instructions" veri setinin bir kısmını yükledik. İnceleyelim:

```python
dataset
```

Yüklediğimiz veri setinde "talimat" alanında sorular, " çıktı" alanında ise cevaplar yer almaktadır.

GPT-2 modelinin eğitimi için soru ve cevapların birbirini takip ederek yer aldığı örneklerden oluşan yeni bir veri seti oluşturalım:

```python
train_dataset = { "example":  [talimat + çıktı for talimat, çıktı in zip(dataset["talimat"], dataset[" çıktı"])]}
train_dataset = Ds.from_dict(train_dataset)

new_dataset = datasets.DatasetDict( {"train":train_dataset} )
new_dataset = new_dataset["train"]

# GPT-2 girdilerini sürekli halde alır. Bu nedenle girdi-çıktı ikilisini bir araya getirerek yeni bir veri seti oluşturuyoruz.
```

```python
new_dataset
```

```python
class MyDataSet(Dataset):
  def __init__(self, tokenizer, data, block_size):
    self.tokenizer = tokenizer
    self.data = data
    self.block_size = block_size

  def __len__(self):
    return len(self.data)

  def __getitem__(self, idx):
    item = self.data[idx]
    tokenized_inputs = self.tokenizer(
        item["example"],
        truncation=True,
        padding="max_length",
        max_length=self.block_size,
        return_tensors="pt"
    )
    return tokenized_inputs
```

```python
data = MyDataSet(tokenizer, new_dataset, 512)
data_collator = DataCollatorForLanguageModeling(tokenizer=tokenizer, mlm=False)
print(dir(data))
```

```python
print(data.data)
print(data.tokenizer)
```

Aşağıda fine-tuning işlemi için gerekli olan parametreler ayarlanmaktadır:

```python
training_args = TrainingArguments(
    per_device_train_batch_size=8,
    num_train_epochs=10,
    learning_rate=5e-5,
    logging_steps=100,
    output_dir='./results'
)

trainer= Trainer(
    model=model,
    args=training_args,
    train_dataset=data,
    eval_dataset=None,
    data_collator=data_collator,
)
```

```python
trainer.train()
```

```python
model.eval()
```

```python
prompt ="Zor günlere rağmen hayatı sevmek için 5 neden sırala."

input_ids = tokenizer(prompt, return_tensors ="pt").input_ids
attention_mask = tokenizer(prompt, return_tensors = "pt").attention_mask

input_ids = input_ids.to(device)
attention_mask = attention_mask.to(device)

input_ids
```

Eğitilen modelle aynı soru için yanıtlar üretelim:

```python
output = model.generate(
    input_ids=input_ids,
    attention_mask=attention_mask,
    pad_token_id=tokenizer.pad_token_id,
    max_length=100,
    num_beams=5,
    temperature=1.5,
    top_k=50,
    do_sample=True
)
```

```python
output
```

```python
text = tokenizer.decode(output[0], skip_special_tokens=True)
print(text)
```
